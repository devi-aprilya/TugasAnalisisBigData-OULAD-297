# **Exploratory Data Analysis**

```python
#untuk menghapus baris yang mengandung nilai null agar menghasilkan data yang akurat
StudentAst.dropna(inplace=True)
StudentInfo.dropna(inplace=True)
StudentVle.dropna(inplace=True)
StudentRgst.dropna(inplace=True)
Vle.dropna(inplace=True)
Courses.dropna(inplace=True)
Ast.dropna(inplace=True)
```

```python
#Menggabungkan antara data student info dan student assesment
df=pd.merge(StudentInfo, StudentAst, how='left', on=['id_student'])
df.info()
```

```python
#Menampilkan presentase gender
#dapat dilihat bahwa gender male lebih banyak daripada female
student_gender=StudentInfo.gender.value_counts()/len(StudentInfo.gender) #menghitung persentase jenis kelamin siswa dari DataFrame
colors = ['#abc4ff', '#9cadce'] #mendefinisikan daftar warna yang akan digunakan dalam plot pie
student_gender.plot.pie(autopct='%.2f', fontsize=12, colors=colors) #membuat plot pie dengan data persentase jenis kelamin yang telah dihitung sebelumnya
plt.title("Precentase gender siswa") #menambahkan judul pada plot pie
```

```python
#menampilkan perbandingan gender female dan man berdasarkan final result
#dari tampilan tabel dibawah dapat dilihat bahwa gender female lebih tinggi dalam pass , fail , distinction dan withdrawn
gender_column = df['gender'] #mengelompokkan data berdasarkan jenis kelamin siswa
result_gender_counts = df.groupby(["final_result", gender_column]).size().unstack(level=1) #mengelompokkan data dalam DataFrame berdasarkan dua kolom, yaitu final result dan gender.
colors = ['#809bce', '#eac4d5'] #mendefinisikan daftar warna yang akan digunakan dalam histogram
result_gender_counts.plot(kind='bar', color=colors) #membuat histogram batang dengan data yang telah Anda siapkan sebelumnya
plt.title("Histogram final result by gender") #menambahkan judul pada histogram batang
```

```python
#Menampilkan histogram dari final result siswa
# Hasil menunjukkan bahwa siswa yang fail hampir setengah dari siswa yang lulus
fig, ax = plt.subplots(figsize=(8, 6)) #membuat sebuah gambar (figure) dan satu set sumbu (axes) dengan ukuran gambar sebesar 8x6 inci
StudentInfo['final_result'].value_counts().plot.bar(color='#abc4ff') #menghitung jumlah kemunculan setiap nilai unik dalam kolom
plt.title("Histogram final result") #menambahkan judul pada histogram batang yang akan dibuat
```

```python
#Menampilkan jumlah siswa dengan berbagai level edukasi
#dapat dilihat bahwa paling banyak siswa dengan A level
fig, ax = plt.subplots(figsize=(8, 6)) #membuat sebuah gambar (figure) dan satu set sumbu (axes) dengan ukuran gambar sebesar 8x6 inci
StudentInfo['highest_education'].value_counts().plot.bar(color='#abc4ff') #menghitung jumlah kemunculan setiap nilai unik dalam kolom tersebut
plt.title("jumlah siswa dengan berbagai level edukasi")
```

```python
#Menampilkan presentase siswa disability
#Dari plot dibawah dapat dilihat presentase siswa dengan disability adalah 10% dari total keseluruhan siswa
student_disability=StudentInfo.disability.value_counts()/len(StudentInfo.disability)
colors = ['#abc4ff', '#9cadce']
student_disability.plot.pie(autopct='%.2f', fontsize=12, colors=colors)
plt.title("Presentase siswa dengan disability")
```

```python
#Menampilkan hasil final result pada disability
disability = df['disability']
result_disability_counts = df.groupby(["final_result", disability]).size().unstack(level=1)
colors = ['#809bce', '#eac4d5']
result_disability_counts.plot(kind='bar', color=colors)
plt.title("Histogram final result by disability")
```

```python
#Menampilkan presentase siswa dengan disability berdasarkan wilayah
#siswa terbanyak dengan disabilty berada di scotland
disability_region=StudentInfo.region.value_counts()/len(StudentInfo.disability)
colors = ['#7ec4cf', '#d7e3fc', '#ccdbfd', '#eac4d5', '#abc4ff', '#ffc09f', '#eac4d5']
disability_region.plot.pie(autopct='%.2f', fontsize=12, figsize=(10, 8), colors=colors)
plt.title("presentase siswa dengan disability berdasarkan wilayah")
plt.show()
```
